```python
#DF METODOS DE ATRACACAO

arquivo_txt = '/content/drive/MyDrive/ANTAQ-DSBD/MetadadosMovimentacao-DICIONARIO/MetadadosAtracacao.txt'

# Listas para armazenar os IDs e Descrições
ids = []
descricoes = []

# Lendo o arquivo linha por linha
with open(arquivo_txt, 'r', encoding='utf-8') as f:
    next(f)  # Ignora a primeira linha (cabeçalho)
    for linha in f:
        partes = linha.strip().split(";", 1)  # Divide a linha na primeira ocorrência de ponto e vírgula
        if len(partes) == 2:  # Garante que existem duas partes após a divisão
            ids.append(partes[0].strip())  # Adiciona o ID, removendo espaços em branco
            descricoes.append(partes[1].strip().strip('"'))  # Adiciona a Descrição, removendo espaços e aspas

# Criando um DataFrame com os dados
df = pd.DataFrame({
    'ID': ids,
    'Descrição': descricoes
})

# Ajustando a largura das colunas para visualização
pd.set_option('display.max_colwidth', 100)  # Define o comprimento máximo das colunas
pd.set_option('display.expand_frame_repr', False)  # Garante que o DataFrame não será quebrado

# Definindo o alinhamento à esquerda
styled_df = df.style.set_properties(**{'text-align': 'left'})

# Exibindo o DataFrame estilizado
styled_df



```
